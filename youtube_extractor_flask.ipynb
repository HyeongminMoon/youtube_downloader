```python
# making flask app for serving mp3 file from youtube url,
# and will make a pitch shift fuction for it

from flask import Flask, render_template, request, redirect, url_for, flash, jsonify, send_file
import pytube
import os
import shutil
import soundfile as sf
import librosa
import librosa.effects
import subprocess
from pydub import AudioSegment

# create sample flask app
app = Flask(__name__)


#     <p>How to use: Paste Youtube link at left slot, if you want to change pitch, use right slot. ex)-2, 5</p>
#     <p>If you check the below box, you can get seperated MR and voice file(recommend this for practicing vocal)</p>
#     <p>Please wait. this will take about 30 seconds.</p>

@app.route('/')
def home(): # return html string
    return '''
    <h1>Youtube Audio Extractor</h1>
    <p> 키 변환 필요하다면 오른쪽 칸에 숫자 ex)-2, 5</p>
    <p> 템포 변환 필요하다면 아랫 칸에 숫자(%) ex)110, 85</p>
    <p> 10~30초정도 걸리니 기다려주세요.</p>
    <br></br>

    <form action="/pitch_shift" method="post" onsubmit="return handle_form_submission()">
        <input type="text" name="url" placeholder="youtube url" style="width:400px;height:50px;font-size:30px;">
        <input type="text" name="pitch_shift" placeholder="pitch_shift" style="width:200px;height:50px;font-size:30px;">
        <br></br>
        <input type="text" name="tempo_shift" placeholder="tempo_shift" style="width:200px;height:50px;font-size:30px;">
        <p> Extract MR & Voice(+20 sec)<input type="checkbox" name="extract_voice" value="extract_voice"> </p>
        
        <button type="submit">Submit</button>
    </form>

    <script>
        function handle_form_submission() {
            button = document.getElementsByTagName('button')[0]
            button.disabled = true
            button.innerHTML = 'Running...'
            return true
        }
    </script>


    '''

@app.route('/pitch_shift', methods=['POST'])
def pitch_shift():
    if request.method == 'POST':
        result = request.form
        url = result['url']
        if url == 'secret':
            # get all file list in directory named covers
            covers = os.listdir('covers')
            
            covers = ['covers/' + cover for cover in covers]
            covers = [cover for cover in covers if cover.endswith('.mp3') or cover.endswith('.mp4')]
            
            return render_template('secret.html', 
                                   covers=covers
                                  )
        
        pitch_shift = result['pitch_shift']
        tempo_shift = result['tempo_shift']
        is_extract_voice = request.form.get('extract_voice')
        is_mp3 = request.form.get('mp3')

        # get wav file from youtube url
        audio_file = get_audio_file(url)
        if audio_file is None:
            return 'Invalid youtube url'

        # shift pitch
        shifted_file = None
        if pitch_shift and int(pitch_shift) != 0:
#             shifted_file = shift_pitch(audio_file, int(pitch_shift))
            shifted_file = hq_shift_pitch(audio_file, int(pitch_shift))
        # shift tempo
        if tempo_shift and int(tempo_shift) != 0:
            if shifted_file:
                shifted_file = change_tempo(shifted_file, float(int(tempo_shift)/100.0))
            else:
                shifted_file = change_tempo(audio_file, float(int(tempo_shift)/100.0))


        # extract voice
        if is_extract_voice:
            if shifted_file:
                instrument, voice = seperator(shifted_file)
            else:
                instrument, voice = seperator(audio_file)
        else:
            instrument, voice = None, None
            
        # convert to mp3
        audio_file = wav_to_mp3(audio_file)
        if shifted_file:
            shifted_file = wav_to_mp3(shifted_file)
        if instrument:
            instrument = wav_to_mp3(instrument)
        if voice:
            voice = wav_to_mp3(voice)

        # remove wav files
        delete_ext_wav()
            
        # flask result page if extracted_voice_file is not None, play it
        return render_template('result.html', 
                               audio_file=audio_file, 
                               shifted_file=shifted_file, 
                               instrument=instrument,
                               voice=voice)

    return 'Invalid request'


@app.route('/<audio_file_name>', methods=['GET'])
def returnAudioFile(audio_file_name):
    # abs
    abs_file_path = os.path.abspath(audio_file_name)
    # abs_file_path = os.path.join(os.getcwd(), audio_file_name)
    
    return send_file(abs_file_path, mimetype='audio/wav')

@app.route('/covers/<audio_file_name>', methods=['GET'])
def returnCover(audio_file_name):
    
    abs_file_path = os.path.join(os.getcwd(), 'covers', audio_file_name)
    
    return send_file(abs_file_path, mimetype='audio/wav')

# get audio file from youtube url using pytube with low quality
def get_audio_file(url):
    # check url is valid youtube url
    if 'youtube' not in url and 'youtu.be' not in url:
        return None
    
    yt = pytube.YouTube(url)
    audio_stream = yt.streams.get_audio_only()
    filename = remove_letters_from_file_name(yt.title)
    out_file = audio_stream.download(filename=filename + '.wav')
    return out_file

# convert mp3 to wav as ffmpeg with quiet option
def audio_convert_ffmpeg(file, ext='wav'):
    if os.path.isfile(os.path.basename(file)[:-3] + ext):
        os.remove(os.path.basename(file)[:-3] + ext)

    # get absoulte path of file
    abs_file_path = os.path.abspath(file)

    # convert mp3 to wav
    # os.system(f'ffmpeg -i "{abs_file_path}" -loglevel panic -ac 1 -ar 16000 "{abs_file_path[:-3]}.wav"')
    subprocess.call(f'ffmpeg -i "{abs_file_path}" -loglevel panic -ac 1 -ar 16000 "{abs_file_path[:-3]}{ext}"', shell=True)

    return os.path.basename(file)[:-3] + ext

# convert wav to mp3 with pydub AudioSegment
def wav_to_mp3(file):
    if os.path.isfile(os.path.basename(file)[:-3] + 'mp3'):
        os.remove(os.path.basename(file)[:-3] + 'mp3')

    # get absoulte path of file
    abs_file_path = os.path.abspath(file)

    # convert wav to mp3
    AudioSegment.from_file(abs_file_path).export(abs_file_path[:-3] + 'mp3', format='mp3')

    return os.path.basename(file)[:-3] + 'mp3'

# pitch shift with librosa
def shift_pitch(file, shift):
    if os.path.isfile(os.path.basename(file)[:-3] + 'pitch_shifted.wav'):
        os.remove(os.path.basename(file)[:-3] + 'pitch_shifted.wav')

    # get absoulte path of file
    abs_file_path = os.path.abspath(file)

    # pitch shift
    y, sr = librosa.load(abs_file_path)
    y_shifted = librosa.effects.pitch_shift(y, sr, n_steps=shift)
    dst_name = abs_file_path[:-3] + f'_shifted{str(shift)}.wav'
    sf.write(dst_name, y_shifted, sr, 'PCM_24')

    return dst_name

import pyrubberband as pyrb
# high quality pitch shift with pyrubberband
def hq_shift_pitch(file, shift):
    y, sr = librosa.load(file)
    y_shifted = pyrb.pitch_shift(y, sr, shift)
    dst_name = file[:-4] + f'_shifted{str(shift)}.wav'
    sf.write(dst_name, y_shifted, sr, format='wav')

    return dst_name

# extract human voice from wav file
def extract_voice(file):
    y, sr = librosa.load(file)
    y_harmonic, y_percussive = librosa.effects.hpss(y)
    dst_name = file + '_voice.wav'
    sf.write(dst_name, y_percussive, sr, format='wav')

    return dst_name

def seperator(file):
    abs_file = os.path.abspath(file)
    
    os.system(f'python /workspace/youtube_download/vocal-remover/inference.py -i {abs_file} --tta --gpu 0')
    return abs_file[:-4] + '_Instruments.wav', abs_file[:-4] + '_Vocals.wav'

# delete ext wav files in current directory
def delete_ext_wav():
    for file in os.listdir():
        if file.endswith('.wav'):
            os.remove(file)

from unittest import TestCase, main

class Test(TestCase):

    def test_hq_shift_pitch(self):
        self.assertEqual(hq_shift_pitch('test.wav', 1), 'test_shifted1.wav')
        self.assertEqual(hq_shift_pitch('test.wav', -1), 'test_shifted-1.wav')

    def test_extract_voice(self):
        self.assertEqual(extract_voice('test.wav'), 'test_voice.wav')

    def test_seperator(self):
        self.assertEqual(seperator('test.wav'), ('test_Instruments.wav', 'test_Vocals.wav'))

import re

# remove letters from file name that are not involved in Korean, English, Numbers
def remove_letters_from_file_name(file_name):
    # remove letters from file name that are not involved in Korean, English, Numbers
    file_name = re.sub(r'[^가-힣a-zA-Z0-9]', '', file_name)
    return file_name

# change tempo with librosa
def change_tempo(file, tempo):
    sr = 44100
    # change tempo
    y, sr = librosa.load(file)
    y_tempo = pyrb.time_stretch(y, sr, tempo)
    dst_name = file[:-3] + f'_tempo{str(tempo)}.wav'
    sf.write(dst_name, y_tempo, sr, format='wav')
    
    return dst_name

# run flask app on port 8010, host=0.0.0.0
if __name__ == '__main__':
    # run unittest
#     main(argv=['first-arg-is-ignored'], exit=False)
    app.run(port=8010, host='0.0.0.0')


```

```python
import numpy as np

# from voice only file, seperate to audio files that envolves one syllable each
# def seperate_to_syllables_files(voice):
voice = 'test_Vocals.mp3'

# from voice only file, seperate to audio files that envolves one syllable each
def seperate_to_syllables_files(voice):
    y, sr = librosa.load(voice)

    # get syllables
    syllables = librosa.effects.split(y, top_db=40)

    # save syllables to files
    for i, syllable in enumerate(syllables):
        start, end = syllable
        syllable_name = voice[:-4] + f'_syllable{i}.wav'
        sf.write(syllable_name, y[start:end], sr, format='wav')

    return syllables


seperate_to_syllables_files(voice)
```
